```python
hr_df_1 = pd.read_excel('HA_Portal Metrics_11-15.xlsx', sheetname=0 )
hr_df_2 = pd.read_excel('HA_Portal Metrics_11-15.xlsx', sheetname=1, skiprows=[0])
```

```python
hr_df_1.head()
```

```python
hr_df_2.head()
```

```python
hr_df = pd.merge(hr_df_1, hr_df_2, how = 'left')
```

```python
hr_df = hr_df.drop('Unnamed: 11', axis = 1)
```

```python
hr_df.head()
```

```python
hr_df['score_total'] = pd.Series(hr_df['Score']/hr_df['Total'])
```

```python
hr_df.head()
```

```python
ttest = stats.ttest_ind
group = list(hr_df['score_total'].groupby(hr_df['Interviewer Name']))
```

```python
p_value = ttest(group[1][1], group[2][1]).pvalue
plt.hist(group[1][1], label = 'Beaker')
plt.hist(group[2][1], label='Mhevans')
#plt.hist(group[1][1], label = 'Beaker')
plt.title('Hypothesis Test Comparing Each Interviewer')
plt.xlabel('Score/Total')
plt.ylabel('Frequency')
plt.legend()
plt.text(x = .45, y = 9, s = 'P-value: ' + str(p_value)[:6], fontsize = 16)
plt.text(x = .34, y = 8, s = '=> not statisticall significant', fontsize = 15)

```

```python
hr_df['Datetime'] = pd.to_datetime(hr_df['Datetime'])

# Make a feature for the day of week
hr_df['dayofweek'] = hr_df['Datetime'].apply(lambda x: x.dayofweek)

# Make a feature for month
hr_df['month'] = hr_df['Datetime'].apply(lambda x: x.month)

# Make a feature for hour
hr_df['hour'] = hr_df['Datetime'].apply(lambda x: x.hour)

hr_df['year'] = hr_df['Datetime'].apply(lambda x: x.year)
```

```python
hr_df.head()
```

```python
hr_df['am_pm'] = pd.Series()
```

```python
for i in range(len(hr_df)):
    if hr_df['hour'][i] > 12:
        hr_df['am_pm'][i] = 1
    else:
        hr_df['am_pm'][i] = 0
        
```

```python
hr_df.head()
```

```python
ttest = stats.ttest_ind
group = list(hr_df['score_total'].groupby([hr_df['am_pm']]))
```

```python
group[0][1]
```

```python
p_value = ttest(group[0][1], group[1][1]).pvalue
plt.hist(group[1][1], label = 'PM')
plt.hist(group[0][1], label='AM', color = 'green')
#plt.hist(group[5][1], label='AM')
#plt.hist(group[1][1], label = 'PM')
plt.title('Hypothesis Test Comparing Time of Day')
plt.xlabel('Score/Total')
plt.ylabel('Frequency')
plt.legend()
plt.text(x = .4, y = 10, s = 'P-value: ' + str(p_value)[:6], fontsize = 16)
plt.text(x = .33, y = 7.5, s = '=> not statisticall significant', fontsize = 15)

```
